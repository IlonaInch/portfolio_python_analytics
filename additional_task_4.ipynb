```python
#считываем excel файл
import pandas as pd

income = pd.read_excel (r'C:/Users/User/Downloads/4_inn.xls')
print (income)
```

```python
income.columns
```

```python
#нам нужны инн и доход, забираем их в датафрэйм income_inn
income_inn= income[['head_inn','income,RUB']]
```

```python
income_inn.head()
```

```python
#назовем текстовый файл "инн"
inn=pd.DataFrame(open(path).readlines())
```

```python
# чтобы в дальнейшем соединить 2 датафрейма, назову колонку  head_inn, также как и в файле эксель
inn.columns=['head_inn']
```

```python
#head_inn иммет тип "объект", поменяем его на int64
inn.dtypes
```

```python
#удалим в каждом инн "\n"
inn['head_inn'] = inn['head_inn'].str.replace(r'\n', '').astype('int64')
```

```python
inn
```

```python
#видим, что тип данных изменился на нужный
inn.dtypes
```

```python
#соединим два тадафрейма по колонке "head_inn"
merge=inn.merge(income_inn, on='head_inn')
```

```python
#считаем сумму доходов тех инн, которые были в текстовом файле
merge['income,RUB'].sum()
```

```python
#сумма всех инн в эксель-файле
summed=income_inn['income,RUB'].sum()
summed
```
